```python
df.drop('No',axis=1,inplace=True)
df.drop('X1 transaction date',axis=1,inplace=True)
```

```python
df.drop('X5 latitude',axis=1,inplace=True)
df.drop('X6 longitude',axis=1,inplace=True)
```

```python
x=df[['X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores']]
y=df[['Y house price of unit area']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
X_train
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model=linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
prediction = model.predict(X_test)
```

```python
accuracy = model.score(X_test, y_test)*100
print(f"The accurancy of the model is {round(accuracy, 2)}%")
```
